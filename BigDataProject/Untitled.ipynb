```python
df_products = pd.read_csv('/home/pirl/와인사업기획/grocery_products.csv',encoding='cp949')
df_products.head()
```

```python
df_reds = df_products.loc[df_products.aisle_id == 28,['product_id','product_name']]
df_reds.head()
```

```python
df_whites = df_products.loc[df_products.aisle_id == 62,['product_id','product_name']]
df_whites.head()
```

```python
df_spaks = df_products.loc[df_products.aisle_id == 134,['product_id','product_name']]
df_spaks.head()
```

```python
from collections import Counter
import re

# p = re.compile('[a-zA-Z]+')
wine_words = []

for wine in df_reds.product_name.values:
    wine_words += map(lambda x: x.lower(),re.findall(r'[A-Za-z]+',wine))
```

```python
for wine in df_whites.product_name.values:
    wine_words += map(lambda x: x.lower(),wine.split())
```

```python
for wine in df_spaks.product_name.values:
    wine_words += map(lambda x: x.lower(),wine.split())
    
```

```python
grape = Counter(wine_words)
```

```python
sorted(grape.items(),key=lambda x:x[1],reverse=True)
```

```python
wine_list = ['cabernet','sauvignon','pinot','noir','zinfandel','merlot',
             'chianti','malbec','shiraz','syrah','zin','sherry','tempranillo',
             'moscato','chardonnay','sirah','carbernet']

# cabernet sauvignon,pinot noir, zinfandel, merlot, chianti, malbec,(shiraz,syrah), sherry, tempranillo
```

```python
df_reds.product_name.values
```

```python
grape_name_wine = set()

for grape in wine_list:
    for wine,index in zip(df_reds.product_name.values,df_reds.product_name.index):
        if(grape in wine.lower()) == True:
            grape_name_wine.add(index)
```
