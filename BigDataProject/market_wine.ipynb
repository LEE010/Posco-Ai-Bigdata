```python
from collections import Counter
import re

# p = re.compile('[a-zA-Z]+')
wine_words = []

for wine in df_reds.product_name.values:
    wine_words += map(lambda x: x.lower(),re.findall(r'[A-Za-z]+',wine))
```

```python
def word_find(values):
    words = []

    for word in values:
        words += map(lambda x: x.lower(),re.findall(r'[A-Za-z]+',word))
    
    return sorted(Counter(words).items(),key=lambda x:x[1],reverse=True)
```

```python
red_words = word_find(df_reds.product_name.values)
white_words = word_find(df_whites.product_name.values)
spak_words = word_find(df_spaks.product_name.values)
```

```python
red_list = ['cabernet','sauvignon','pinot','noir','zinfandel','merlot',
             'chianti','malbec','shiraz','syrah','zin','sherry','tempranillo',
             'moscato','chardonnay','sirah','carbernet']

# cabernet sauvignon,pinot noir, zinfandel, merlot, chianti, malbec,(shiraz,syrah), sherry, tempranillo
```

```python
def wine_searcher(df_wines,target_wine_list):
    grape_name_wine = set()

    for grape in target_wine_list:
        for wine,index in zip(df_wines.product_name.values,df_wines.product_name.index):
            if(grape in wine.lower()) == True:
                grape_name_wine.add(index)
    return list(grape_name_wine)
```

```python
res_reds = wine_searcher(df_reds, red_list)
```
