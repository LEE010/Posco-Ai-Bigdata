```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
data = pd.read_csv('./상품구매.csv')
data.head()
```

```python
data.ID.value_counts()
```

```python
data.PRODUCT.value_counts()
```

```python
df_pre = ~data.pivot(index = "ID",columns="PRODUCT",values="PRODUCT").isnull()
df_pre
```

```python
# min_support = 0.5

df_freq = apriori(df_pre,min_support=0.5,use_colnames=True)
df_freq.round(3)
```

```python
df_asso_rule = association_rules(df_freq,metric="confidence",min_threshold=0.5)
df_asso_rule.round(3)
```
