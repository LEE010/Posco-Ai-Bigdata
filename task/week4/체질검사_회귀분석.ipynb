```python
df_fat_y = df_fat["FAT"]
df_fat_x = df_fat.drop("FAT", axis=1, inplace=False)
df_vif = pd.DataFrame()
df_vif["variable"] = df_fat_x.columns
df_vif["vif"] = [variance_inflation_factor(df_fat_x.values,i) for i in range(df_fat_x.shape[1])]
df_vif.sort_values("vif", inplace=True)
df_vif.round(3)
```

다중공선성 검정 결과 모든 값이 높은 결과를 나타냈다.

```python
df_fat_y = df_fat["FAT"]
df_fat_x = df_fat.drop(["FAT","HIP","WRIST","CHEST","KNEE"], axis=1, inplace=False)
df_vif = pd.DataFrame()
df_vif["variable"] = df_fat_x.columns
df_vif["vif"] = [variance_inflation_factor(df_fat_x.values,i) for i in range(df_fat_x.shape[1])]
df_vif.sort_values("vif", inplace=True)
df_vif.round(3)
```

공선성이 큰 항목부터 하나씩 제거한 결과 그나마 적어진 결과를 얻을 수 있었다.

```python
df_fat_x_const = sm.add_constant(df_fat.drop(["FAT","HIP","WRIST","CHEST","KNEE"], axis=1, inplace=False))
reg_multi_model = sm.OLS(df_fat_y, df_fat_x_const)
reg_multi_res = reg_multi_model.fit()
print(reg_multi_res.summary())
```

p-value가 유의수준보다 매우 높은 결과를 보이는 항목들을 제거한다
>* THIGH: 허벅지 둘레
>* BICEPS: 이두근 둘레
>* FOREARM: 팔뚝 둘레
>> 주로 근육으로 키울 수 있는 둘레에 대하여 **p-value가 높게 나타난 것으로 보인다**

```python
reg_multi_model = sm.OLS(df_fat_y, df_fat_x_const.drop(["BICEPS"],axis=1,inplace=False))
reg_multi_res = reg_multi_model.fit()
print(reg_multi_res.summary())
```

```python
reg_multi_model = sm.OLS(df_fat_y, df_fat_x_const.drop(["BICEPS","FOREARM"],axis=1,inplace=False))
reg_multi_res = reg_multi_model.fit()
print(reg_multi_res.summary())
```

```python
reg_multi_model = sm.OLS(df_fat_y, df_fat_x_const.drop(["THIGH","BICEPS","FOREARM"],axis=1,inplace=False))
reg_multi_res = reg_multi_model.fit()
print(reg_multi_res.summary())
```

AGE와 BMI의 p-value가 유의수준 보다 크게 나왔음을 알수있다.  
>* BMI지수는 키와 몸무게만 이용하여 구하는 값이다. 따라서 그 사람의 근골격량을 고려하지 않은 수치라 유의할정도의 영향력은 생기지 않는것 같다. 

```python
reg_multi_model = sm.OLS(df_fat_y, df_fat_x_const.drop(["THIGH","BICEPS","FOREARM","AGE"],axis=1,inplace=False))
reg_multi_res = reg_multi_model.fit()
print(reg_multi_res.summary())
```

```python
reg_multi_model = sm.OLS(df_fat_y, df_fat_x_const.drop(["THIGH","BICEPS","FOREARM","AGE","BMI"],axis=1,inplace=False))
reg_multi_res = reg_multi_model.fit()
print(reg_multi_res.summary())
```

```python
df_reg_coef = pd.DataFrame({"Coef":reg_multi_res.params.values[1:]},index=reg_multi_model.exog_names[1:])
df_reg_coef.plot.barh(y="Coef",legend=False)
```

```python
df_fat_x_scaled = scale(df_fat_x_const.drop(["THIGH","BICEPS","FOREARM","AGE","BMI"],axis=1,inplace=False))
reg_scaled_model = sm.OLS(df_fat_y,df_fat_x_scaled)
reg_scaled_res = reg_scaled_model.fit()
df_reg_scaled_coef = pd.DataFrame({"Coef": reg_scaled_res.params.values[1:]},index=reg_multi_model.exog_names[1:])
df_reg_scaled_coef.plot.barh(y="Coef",legend=False)
```

스케일 조정 후 확인 결과 복부 둘레를 제외하고 큰 영향력이 없는 것으로 확인된다.
발목 두께와, 목 둘레는 약한 음의 상관관계를 보였는데, 살 찌기 어려운 부위인 만큼 두꺼울 경우 근골격이 기본적으로 좋아 비만도가 적게 나올 것 이라고 추측된다.

```python
def draw_multivarient_plot(dataset, rows, cols, plot_type):
    column_names=dataset.columns.values
    number_of_column=len(column_names)
    fig, axarr=plt.subplots(rows,cols, figsize=(5*cols,5*rows))

    counter=0
    for i in range(rows):
        for j in range(cols):
            if 'violin' in plot_type:
                sns.violinplot(y=column_names[counter],data=dataset, ax=axarr[i][j])
            elif 'box'in plot_type :
                sns.boxplot( y=column_names[counter],data=dataset, ax=axarr[i][j])
            elif 'point' in plot_type:
                sns.pointplot(y=column_names[counter],data=dataset, ax=axarr[i][j])
            elif 'bar' in plot_type:
                sns.barplot(y=column_names[counter],data=dataset, ax=axarr[i][j])

            counter+=1
            if counter==(number_of_column-1,):
                break
```
