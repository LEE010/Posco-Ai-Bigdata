```python
x = 135
v2 = stats.norm.cdf(x, 115, 20)
```

```python
print("{0}분 이상 걸리는 as작업비율: {1: .1f}%".format(x,(1 - v2)*100))
```

```python
v3 = stats.norm.sf(x, 115, 20)
print("{0}분 이상 걸리는 as작업비율: {1: .1f}%".format(x, v3*100))
```

```python
inverse = stats.norm.ppf(v2, loc = 115, scale = 20)
inverse
```

pdf: probability density function 확률밀도함수

cdf: cumulative distribution function 누적분포함수

ppf: inverse cumulative distribution function 누적분포함수의 역함수

rvs: randmo variable sampling 랜덤 표본 생성

```python
xx = np.linspace(0, 100, 400)

for df in [1,3,5,25,50,100,10000]:
    chi2_df = stats.chi2(df=df)
    plt.plot(xx, chi2_df.pdf(xx))
```

```python
# seed 선택을 하여 매번 실행시 동일한 값이 나오도록 설정
np.random.seed(seed = 1234)
# 자유도: 40, 데이터수 : 1000개의 카이제곱 분포를 따르는 데이터 생성
chisq_df40 = np.random.chisquare(df = 40, size = 1000)
# 히스토그램, kde:카이제곱 분포의 kde 생성
sns.distplot(chisq_df40, kde=True)
```

## 실습

```python
chisq = 10
df = 30
# 누적확률 값 계산
prop = stats.chi2.cdf(chisq, df)
print("P(X <={0}):{1:.4f}".format(chisq, prop))
```

```python
chisq = 10
df = 30
# 누적확률 값 계산
prop = stats.chi2.cdf(chisq, df)
prop2 = stats.chi2.sf(chisq, df)
prop3 = stats.chi2(df=df).ppf(prop)
print("P(X <={0}):{1:.4f}".format(chisq, prop))
print("P(X >{0}):{1:.4f}".format(chisq, prop2))
print("P(X <={0}):{1:.4f}".format(chisq, prop3))
```

표본 분산의 비율의 분포

```python
# seed 선택을 하여 매번 실행 시 동일한 값이 나오도록 설정
np.random.seed(seed=1234)
xx = np.linspace(0,3,50)

for df in [1, 3, 5, 25, 50]:
    f_df = stats.f(dfn=df, dfd=10)
    plt.plot(xx, f_df.pdf(xx))
```

```python
# seed 선택을 하여 매번 실행 시 동일한 값이 나오도록 설정
np.random.seed(seed=1234)
# 분자자유도 10, 분모자유도 10
# 데이터 수: 1000개의 F분포를 따르는 데이터 생성
F_df10_10 = np.random.f(dfnum = 10, dfden=10, size=1000)
sns.distplot(F_df10_10, kde=True)
```

## 포아송분포

pmf: Probability mass function 확률 질량 함수:(이산일경우)
pdf: 연속일경우 => 확률 밀도 함수

```python
n=3
for i in range(n+1):
    # 이항분포 pmf
    prop = stats.binom.pmf(k=i, n=n, p=0.4)
    print("P(X ={0}):{1:.3f}".format(i,prop))
```

```python
for n in [3,5,25,50]:
    binorm_df=stats.binom(n=n, p=0.4)
    xx = np.linspace(0,n,n+1)
    plt.plot(xx,binorm_df.pmf(xx))
```

```python
for mu in [2,4,6,10]:
    poisson_df = stats.poisson(mu=mu)
    xx = np.linspace(0,3*mu,3*mu+1)
    plt.plot(xx,poisson_df.pmf(xx))
```

```python
mu = 2
# 1분당 최대 2회 전화가 걸려올 확률
prop = stats.poisson.pmf(3,mu)
# 1분당 최대 2회 이하의 전화가 걸려올 확률
cdf_prop = stats.poisson.cdf(2,mu)
print('1분당 {0}번의 전화가 걸려올 확률: {1:.4f}'.format(3, prop))
print('1분당 {0}번 이하의 전화가 걸려올 확률: {1:.4f}'.format(2, cdf_prop))
```

# 구간추정

```python
ds_example = pd.read_csv('./files/실습화일/EXH_QC.csv', engine='python')
```

```python
df = ds_example[['Cabbage weight']]
lower, upper = stats.norm.interval(0.95, loc=2.7, scale=0.397/np.sqrt(40))
print('신뢰구간:({0}, {1})'.format(lower.round(2), upper.round(2)))
```

```python
df = pd.DataFrame({"sample": [54.1, 53.3, 56.1, 55.7, 54.0, 54.1, 54.5, 57.1, 55.2, 53.8, 54.1, 54.1, 56.1, 55.0, 55.9, 56.0, 54.3, 53.9, 55.0]})
```

```python
lower, upper = stats.t.interval(0.95, len(df)-1, loc = np.mean(df), scale=stats.sem(df))
print('신뢰구간:({0}, {1})'.format(lower.round(2), upper.round(2)))
```

```python
df = pd.DataFrame({"sample":[74.5, 81.2, 73.8, 82.0, 76.3, 75.7, 80.2, 72.6, 77.9, 82.8]})
t_test = stats.ttest_1samp(df, 76.7)
t,p = t_test.statistic.round(3), t_test.pvalue.round(3)
print('1-sample t -test')
print('t: ',t)
print('p: ',p)
```

```python
df1 = pd.DataFrame({"sample": [6,5,5,4,6,7,6,4,5,6,4,5,5,6,4,8,6,5,6,7]})
df2 = pd.DataFrame({"sample": [7,5,7,8,7,8,8,5,7,6,5,5,6,6,5,7,9,7,7,8]})

t_test = stats.ttest_ind(df1, df2)
t,p = t_test.statistic.round(3), t_test.pvalue.round(3)
print('2-sample t -test')
print('t: ',t)
print('p: ',p)
```

```python
df1 = pd.DataFrame({'A': [829, 623, 721, 678, 723, 790, 720, 670, 690, 711]})
df2 = pd.DataFrame({'B': [721, 821, 910, 678, 723, 790, 711, 790, 745, 891]})
df3 = pd.DataFrame({'C': [621, 915, 888, 721, 894, 834, 841, 912, 845, 889]})

f_test = stats.f_oneway(df1, df2, df3)
f,p = f_test.statistic.round(3), t_test.pvalue.round(3)
print('one way ANOVA')
print('f: ',f)
print('p: ',p)
```

```python
count = 15
nobs = 100
value = 0.1
stat, pval = proportions_ztest(count, nobs, value)
print("1 proportion test")
print("z: %.3f"% stat)
print("z: %.3f"% pval)
```

```python
count = np.array([4,1])
nobs = np.array([1000, 1200])
stat, pval = proportions_ztest(count, nobs)
print("2 proportion test")
print("z: %.3f"% stat)
print("p: %.3f"% pval)
```
