## 데이터 결합

```python
df1 = pd.DataFrame(columns=['A','B','C','D'])
df2 = pd.DataFrame(columns=['A','B','C','D'])
df3 = pd.DataFrame(columns=['A','B','C','D'])
df4 = pd.DataFrame(columns=['A','B','C','D'])
```

```python
list(enumerate([5,6,7,8]))
```

```python
def add_item(df,index):
    keys = df.columns
    for k in range(index*4, (index+1)*4):
        df.loc[k] = [ keys[0][0]+str(k), keys[1][0]+str(k), keys[2][0]+str(k), keys[3][0]+str(k)]
```

```python
for i,df in enumerate((df1, df2, df3, df4)):
    add_item(df, i)
```

```python
print(df1, df2, df3, df4)
```

```python
pd.concat([df1,df2,df3,df4])
```

```python
pd.concat([df1,df2,df3,df4],axis=1)
```

```python
pd.concat([df1,df2,df3,df4], join='inner',axis =1)
```
